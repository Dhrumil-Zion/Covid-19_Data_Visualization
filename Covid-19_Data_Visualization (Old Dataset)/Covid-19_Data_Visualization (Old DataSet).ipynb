```python
World_Data=pd.read_csv(r"covid_19_data.csv")


```

```python
World_Data.rename(columns={'ObservationDate':'Date','Province/State':'State','Country/Region':'Region',
                          'Last Update':'l_update'},inplace=True)
```

```python
World_Data
```

```python
World_Data.query('Region=="India"').head()
```

```python
Confirmed = World_Data.groupby('Date').sum()['Confirmed'].reset_index()
Recovered = World_Data.groupby('Date').sum()['Recovered'].reset_index()
Deaths = World_Data.groupby('Date').sum()['Deaths'].reset_index()
```

```python
Confirmed.head()
#Recovered
#Deaths
```

```python
fig= go.Figure()

fig.add_trace(go.Scatter(x=Confirmed['Date'],y=Confirmed['Confirmed'],
             mode="lines+markers",name="Confirmed",line=dict(color="blue",width=2)))

fig.add_trace(go.Scatter(x=Deaths['Date'],y=Deaths['Deaths'],
             mode="lines+markers",name="Deaths",line=dict(color="red",width=2)))

fig.add_trace(go.Scatter(x=Recovered['Date'],y=Recovered['Recovered'],
             mode="lines+markers",name="Recovered",line=dict(color="green",width=2)))
```

```python
World_Data.head()
```

```python
time_series= pd.read_csv(r"time_series_covid_19_confirmed.csv")
```

```python
#We have to merge the data frame World_Data and time-series which are going to generate
#We have to make first cols same in both data frames so Worl_data have Provience/states = states and c/region = region 

time_series.rename(columns={'Province/State':'State','Country/Region':'Region'},inplace=True)
```

```python
time_series

time_series_with_lat_long = pd.merge(World_Data,time_series,on=['Region','State'])
```

```python
time_series_with_lat_long
```

```python
fig=px.density_mapbox(time_series_with_lat_long,lat="Lat",
                      lon="Long",
                      hover_name="Region",
                      hover_data=["Confirmed","Deaths","Recovered"],
                      animation_frame="Date",
                      color_continuous_scale="Portland",
                      radius=7,
                      height=700,zoom=0)

fig.update_layout(title="World Data Time-line")

fig.update_layout(mapbox_style="open-street-map",mapbox_center_lon=0)

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

```
